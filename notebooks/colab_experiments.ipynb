## 📈 Summary

The experiments compare Gaussian activation functions against ReLU, GELU, and Swish on:
- Simple MLP networks
- LeNet-5 (classic CNN)
- VGG-Mini (deep CNN)
- ResNet-Mini (residual network)

### Parameters tested:
- `μ = 0, σ = 1` (standard Gaussian)
- `μ = 0.5, σ = 1` (shifted Gaussian)
- `μ = 0, σ = 0.5` (narrow Gaussian)
