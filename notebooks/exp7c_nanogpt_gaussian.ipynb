# 🧪 Experiment 7c: nanoGPT + LearnableGaussian (Optimized)

**优化版本** - 初始平缓、有平移、分层学习率

## 优化策略

1. **初始平缓**: sigma=2.0 (大 sigma，便于梯度传播)
2. **有一定平移**: mu=0.5 (让网络学习最优位置)
3. **残差缩放**: 可学习的残差连接缩放因子
4. **分层学习率**: Gaussian 参数使用 1.5x 学习率 (lr=5e-4)
5. **Warmup**: 100 步 warmup 阶段
