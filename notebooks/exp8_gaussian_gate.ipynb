# 🧪 Experiment 8: nanoGPT — GELU vs GaussianGate

## 核心思路

用高斯函数作为**软门控**，而不是直接替换激活函数：

```
GaussianGate(x) = exp(-(x - mu)² / (2σ²)) * x
```

类比 GELU：`GELU(x) = sigmoid(1.702x) * x`

| | GELU | GaussianGate |
|---|---|---|
| 门控函数 | sigmoid(1.702x) | exp(-(x-mu)²/2σ²) |
| 可学习参数 | 0 | 2 (mu, sigma) |
| 门控范围 | 单调递增 | 局部响应 |
