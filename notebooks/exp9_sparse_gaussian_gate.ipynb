# 🧪 Experiment 9: SparseGaussianGate — 模拟神经元的离散感受野

## 核心思路

用 **N 个高斯的混合** 作为门控，模拟神经元的离散响应模式：

```
gate(x) = Σ w_i * exp(-(x - μ_i)² / (2σ_i²))
output  = gate(x) * x
```

### 神经科学类比

| 概念 | 含义 |
|------|------|
| 每个高斯 | 一个「感受野中心」 |
| μ_i | 感受野位置（离散分布） |
| σ_i | 感受野宽度 |
| w_i | 该感受野的重要性权重 |

### 参数对比

| 激活函数 | 可学习参数 |
|----------|------------|
| GELU | 0 |
| GaussianGate (N=1) | 2 |
| **SparseGaussianGate (N=4)** | 12 |
| **SparseGaussianGate (N=8)** | 24 |
